# Group by and Aggregating in Pandas 


```python
import pandas as pd

df = pd.read_csv(r'C:\Users\ADMIN\Downloads\Pandas files\Flavors.csv')

df

```

```python
# Group by (Similar values, duplicate values)
```

```python
df.groupby('Base Flavor')
```

```python
group_by_frame = df.groupby('Base Flavor')

group_by_frame.mean('Base Flavor')
```

```python
df.groupby('Base Flavor').mean('Base Flavor')
```

```python
## Popular Agregation

Actual rows that were aggregated
```

```python
df.groupby('Base Flavor').count()
```

```python
df.groupby('Base Flavor').min()
```

```python
df.groupby('Base Flavor').max()
```

```python
df.groupby('Base Flavor').sum()
```

```python
### Aggregate function
```

```python
df.groupby('Base Flavor').agg({'Flavor Rating': ['mean','max','count','sum'] })
```

```python
df.groupby(['Base Flavor','Liked']).mean('Base Flavor','Liked')
```

```python
df.groupby(['Base Flavor','Liked']).agg({'Flavor Rating':['mean','max','count','sum']})
```

```python
df.groupby('Base Flavor').describe()
```
